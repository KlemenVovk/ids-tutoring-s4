```python
import sqlite3
import os
from pprint import pprint

CREATE_DB_SCRIPT = "dummy_data/createdb.sql"
POPULATE_DB_SCRIPT = "dummy_data/populatedb.sql"
DB_PATH = "dummy.db"

if os.path.exists(DB_PATH):
    print("Removing old database...")
    os.remove(DB_PATH)

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()

with open(CREATE_DB_SCRIPT, "r") as f:
    print("Creating database...")
    cur.executescript(f.read())

with open(POPULATE_DB_SCRIPT, "r") as f:
    print("Populating database...")
    cur.executescript(f.read())

conn.commit()
```

```python
# A helper function that executes a given query and nicely prints the result
def execute_and_print(query):
    print(query)
    cur.execute(query)
    pprint(cur.fetchall())
    conn.commit()
```

```python
execute_and_print("SELECT * FROM courses")
```

```python
execute_and_print("SELECT * FROM students")
```

```python
execute_and_print("SELECT * FROM instructors")
```

```python
execute_and_print("SELECT * FROM enrollments")
```

### List the courses along with their professors

Let's do a join manually first.

```python
execute_and_print("""
SELECT courses.name, instructors.first_name, instructors.last_name
FROM courses, instructors
WHERE courses.instructor_id = instructors.instructor_id
""")
```

Now let's use the builtin INNER JOIN. Note this is the default type of join so you can write just 'JOIN' and it will also work.

```python
execute_and_print("""
SELECT courses.name, instructors.first_name, instructors.last_name
FROM courses
INNER JOIN instructors ON courses.instructor_id = instructors.instructor_id;
""")
```

### All students enrolled in Machine Learning (names, years, grades)


```python
execute_and_print("""
SELECT s.first_name, s.last_name, e.year, e.grade
FROM students s
INNER JOIN enrollments e ON s.student_id = e.student_id
INNER JOIN courses c ON e.course_id = c.course_id
WHERE c.name = 'Machine Learning'
ORDER BY e.year DESC;
""")
```

### Average grade for each course

```python
execute_and_print("""
SELECT c.name, AVG(e.grade) as average_grade
FROM courses c
INNER JOIN enrollments e ON c.course_id = e.course_id
WHERE e.grade IS NOT NULL
GROUP BY c.name
ORDER BY average_grade DESC;
""")
```

### Average grade for each course per year

```python
execute_and_print("""
SELECT c.name, e.year, AVG(e.grade) as average_grade
FROM courses c
INNER JOIN enrollments e ON c.course_id = e.course_id
WHERE e.grade IS NOT NULL
GROUP BY c.name, e.year
ORDER BY c.name, e.year;
""")
```

### Total number of enrollements for each course

```python
execute_and_print("""
SELECT c.name, COUNT(e.enrollment_id) as num_students
FROM courses c
INNER JOIN enrollments e ON c.course_id = e.course_id
GROUP BY c.name
ORDER BY c.name;
""")
```

### Courses that have had at least 250 enrollments

HAVING works exclusively on aggregate functions and it has to be used after group by.

```python
execute_and_print("""
SELECT c.name, COUNT(e.enrollment_id) as num_students
FROM courses c
INNER JOIN enrollments e ON c.course_id = e.course_id
GROUP BY c.name
HAVING num_students >= 250
ORDER BY c.name;
""")
```

### Students that can still pick Machine Learning (they haven't been enrolled yet).

```python
execute_and_print("""
SELECT s.first_name, s.last_name
FROM students s
WHERE s.student_id NOT IN (
    SELECT e.student_id
    FROM enrollments e
    INNER JOIN courses c ON e.course_id = c.course_id
    WHERE c.name = 'Machine Learning'
)
ORDER BY s.last_name, s.first_name;
""")
```

### New courses - no student enrollments yet

Nested query:

```python
execute_and_print("""
SELECT c.name
FROM courses c
WHERE c.course_id NOT IN (
    SELECT e.course_id
    FROM enrollments e
)
ORDER BY c.name;
""")
```

We can use an outer join as the right tool for the job here

```python
execute_and_print("""
SELECT c.name
FROM courses c
LEFT OUTER JOIN enrollments e ON c.course_id = e.course_id
WHERE e.course_id IS NULL
ORDER BY c.name;
""")
```

### Delete courses with no enrollments

```python
# delete courses with no students
execute_and_print("""
DELETE FROM courses
WHERE course_id IN (
    SELECT c.course_id
    FROM courses c
    LEFT OUTER JOIN enrollments e ON c.course_id = e.course_id
    WHERE e.course_id IS NULL
);
""")
```

### Set the grade of all enrollments that haven't passed yet, to 5.

```python
execute_and_print("""
UPDATE enrollments
SET grade = 5
WHERE grade IS NULL;
""")
```
